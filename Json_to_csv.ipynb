```python
import pandas as pd 
import json
```

T20_wc_match_results

```python
with open('T20 JSON File/t20_wc_match_results.json') as f:
    data = json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head(11)
```

```python
df_match.shape
```

```python
df_match.rename({'scorecard':'match_id'},axis= 1,inplace= True)
df_match.head(11)
```

```python
match_ids_dict = {}

for index, row in df_match.iterrows():
    key1 = row["team1"] + ' Vs ' + row['team2']
    key2 = row["team2"] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row["match_id"]
    match_ids_dict[key2] = row["match_id"]
    
print(match_ids_dict)
```

```python
df_match.to_csv('T20_CSV_File/t20_wc_match_results.csv',index= True)
```

T20_wc_batting_summary.json

```python
with open('T20 JSON File/t20_wc_batting_summary.json') as f:
    data = json.load(f)
    
    all_records = []
    
    for rec in data:
        all_records.extend(rec['battingSummary'])
        


df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
df_batting["OUT/NOT_OUT"] = df_batting.dismissal.apply(lambda x: "OUT" if len(x) > 0 else "NOT_OUT" )
df_batting.head(11)
```

```python
df_batting["batsmanName"]=df_batting["batsmanName"].apply(lambda x:x.replace('â€',''))
df_batting["batsmanName"]=df_batting["batsmanName"].apply(lambda x:x.replace('(c)',''))
df_batting.tail(11)
```

```python
df_batting["mach_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head(11)
```

```python
df_batting.to_csv('T20_CSV_File/t20_wc_batting_summary.csv',index= True)
```

BowlingSummary

```python
with open('T20 JSON File/t20_wc_bowling_summary.json') as f:
    data = json.load(f)

df_bowling = pd.DataFrame(data[0]['bowlingSummary'])
df_bowling.head(11)
```

 Process Players Information

```python
with open('T20 JSON File/t20_wc_player_info.json') as f:
    data = json.load(f)
    
df_player = pd.DataFrame(data)
df_player.head(10)
```

```python
df_player ['name'] = df_player ['name'].apply(lambda x: x.replace('â€', ''))
df_player ['name'] = df_player ['name'].apply(lambda x: x.replace('†', ''))
df_player ['name'] = df_player ['name'].apply(lambda x: x.replace('\xa0', ''))
df_player.head(10)
```

```python
df_player.to_csv('T20_CSV_File/t20_wc_player_info.csv',index= True)
```
